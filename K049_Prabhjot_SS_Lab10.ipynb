Expt. No. 10: Filtering high frequency noise

Aim: Denoising the high frequency noise from the received low frequency signal using a ideal low pass filter.

Theory:<br>
Noise is a unwanted signal which affects the desired signal.The unwanted noise (signal) arises from a number of sources (external and internal).

Denoising of high frequency signal is done using a LOW PASS FILTER

IDEAL LOW PASS FILTER:
A low-pass filter is one that allows low frequencies to pass while attenuating anything above a specified cutoff frequency $ f_c$:

Capture.PNG

import the required libraries

```python
from pylab import*
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
## 1. generate the low frequency sinusoidal filter
t = linspace(-2,2,100)
x = 4*sin(2*pi*1.5*t)
subplot(3,1,1)
plot(t,x)
title("Transmitted Signal")
xlabel('t')
ylabel('x(t)')
show()

n=2*sin(2*pi*8*t)+(3*sin(2*pi*14*t))

subplot(3,1,2)
plot(t,n)
title("Channel Noise")
xlabel("t")
ylabel("n(t)")
show()

Rx=x+n
subplot(3,1,3)
plot(t,Rx)
title("Received Signal")
xlabel('t')
ylabel('xn(t)')
show()

# 5. find the output in time domain by taking inverse fourier transform of Y(w)
# 6. PLOT OF Original SIGNAL
# 7. PLOT OF NOISY AND FILTERED SIGNAL
# 8. plot of frequency spectrum of original signal (i.e S(w))
# 9. plot of frequency spectrum of noisy received signal
# 10. plot of frequency spectrum of recovered signal (i.e Y(w))
```

```python
# 2. The transmitted signal is corrupted by high frequecy noise

w=linspace(0,len(Rx),len(Rx))
X=np.fft.fft(x)
subplot(2,1,1)
plot(w,abs(X))
title("Frequency Spectrum of Original SIgnal")
xlabel("w")
ylabel("X(w)")
show()

subplot(2,1,2)
Rxw=np.fft.fft(Rx)
plot(w,abs(Rxw))
title("Received Signal")
xlabel("t")
ylabel("")
show()
```

```python
# 3. design a ideal low pass filter
fc = 2
wc = 2 * pi * fc
Uw = 1 * ( w > 0)
Udw = 1 * ((w - wc) > 0)
Hw = Uw - Udw
plot(w, Hw)
title("Frequency Spectrum of System Function")
xlabel("w")
ylabel("H(w)")
```

```python
# 4. generate the output by multipling the Y(w) = X(w) . H(w)
Yw = X * Hw
subplot(3,1,1)
plot(w,abs(Yw))
title("Frequency Spectrum of Filtered Signal")
show()

y = np.fft.fft(Yw)
subplot(3,1,2)
plot(t,y)
title("Filtered Signal")
xlabel("t")
ylabel("y(t)")
show()

subplot(3,1,3)
plot(t,x)
title("Original Signal")
xlabel("t")
ylabel("x(t)")
```
